# Data Wrangling

```python
# Clean data
df = pd.read_csv(pre_path + "2019_Yellow_Taxi_Trip_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['store_and_fwd_flag'].unique()
```

```python
mask = df.columns.str.contains('id$|store_and_fwd_flag', regex=True)
drop_cols = df.columns[mask]
drop_cols
```

```python
df = df.drop(columns=drop_cols)
df.head()
```

# Rename

```python
df = df.rename(
    columns={
        'tpep_pickup_datetime': 'pickup_dt',
        'tpep_dropoff_datetime': 'dropoff_dt'
    }
)
df.columns
```

```python
df.dtypes
```

```python
df['pickup_dt'].apply(pd.to_datetime)
```

```python
sel_cols = df.columns[df.columns.str.contains("dt$")]
sel_cols
```

```python
df.loc[:, sel_cols] = df[sel_cols].apply(pd.to_datetime)
```

```python
df.dtypes
```

# Feature Engineering

```python
df.columns
```

```python
df['elapsed_time'] =  df['dropoff_dt'] - df['pickup_dt']
df['speed'] = df['trip_distance'] / df['elapsed_time'].dt.total_seconds() / 60 / 60
df
```

```python
df.sort_values(['speed', 'total_amount'], ascending=[False, True]).dropna(subset=['speed'])
```

```python
df_2019 = df[df['year'] == 2019]
df_2019['month'] = df_2019['date'].dt.month
df_2019
```
