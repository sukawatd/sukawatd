# Session 2: Data Wrangling, Cleansing
Data dictionary of NYC `2019_Yellow_Taxi_Trip_Data.csv` 
https://www1.nyc.gov/assets/tlc/downloads/pdf/data_dictionary_trip_records_yellow.pdf

```python
df = pd.read_csv(pre_path + "2019_Yellow_Taxi_Trip_Data.csv")
```

```python
df.head()
```

```python
df.describe().round(2)
```

```python
df.columns
```

```python
df['store_and_fwd_flag'].value_counts()
```

```python
drop_cols = df.columns[df.columns.str.contains('id$|store_and_fwd_flag')]
drop_cols
```

```python
df_clean = df.drop(columns=drop_cols)
```

```python
df_clean =df_clean.rename(columns={
    "tpep_pickup_datetime": "pickup_dt",
    "tpep_dropoff_datetime": 'dropoff_dt'
})
```

## Change Dtypes

```python
df_clean.dtypes
```

```python
sel_cols = ['pickup_dt', 'dropoff_dt']
df_clean[sel_cols] = (df_clean[sel_cols].apply(pd.to_datetime))
```

```python
df_clean['pickup_dt'].dt.year
```

## Feature Engineering

```python
df_clean['elapsed_time'] = df_clean['dropoff_dt'] - df_clean['pickup_dt']
# V = S/T

df_clean['avg_speed'] = df_clean['trip_distance'] / df_clean['elapsed_time'].dt.total_seconds() / 60 / 60
df_clean.head(3)
```

## Barplot

```python
df_p = df
df_p['month'] = df_p['date'].dt.month
df_p = df_p.pivot_table(
    index='month',
    columns='year',
    values='travelers'
)
df_p.head(2)
```

```python
sns.heatmap(df_p, cmap='Blues', annot=True, fmt='.10g')
```

```python
ax = df_p.plot(kind='bar', ylabel='travelers')

ax.legend(title='', loc='center', bbox_to_anchor=(0.5, -0.3), ncol=3, frameon=False)
```

```python
1e6
```

## Plot Distributions

```python
df.query('year == 2019').plot(
    kind = 'hist',
    x = 'date',
    y = 'travelers',
    bins=30,
    figsize=(12, 5)
)
```

```python
pivot_df = df.pivot_table(columns='year', values='travelers', index='date')

fig, ax = plt.subplots(3, 1, figsize=(10, 6), dpi=80)

ax = pivot_df.plot(kind='hist', subplots=True, bins=20, ax=ax, sharex=True, density=True)
pivot_df.plot(kind='kde', subplots=True, ax = ax, sharex=True, color='k')
```

## Groupby
